```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
sample_size = 10**2
```

```python
nums_means = [10**1, 10**2, 10**3]
```

```python
lo_lim, hi_lim = -4, 4
```

```python
plt.style.use('seaborn-white')
```

```python
xs = np.linspace(lo_lim, hi_lim, sample_size)
```

```python
fig, ax = plt.subplots(nrows=len(nums_means), sharex=True, sharey=True, figsize=(8,15))

for i, num_means in enumerate(nums_means):
    means = [np.random.normal(size=sample_size).mean() for _ in range(num_means)]

    ax[i].set_title('{} means: Centered at {:.4f}'.format(num_means, np.mean(means)), fontsize=15)

    ax[i].plot(xs, norm.pdf(xs), label='Std. Normal Distribution', linewidth=0.80)
    ax[i].hist(means, alpha=0.5, weights=np.ones_like(means)/num_means, label='Distributions of Means')

    ax[i].legend()

fig.savefig('distribution-of-means.png')
```
