```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
num_vars = 10**2
```

```python
sample_sizes = [10**2, 10**3, 10**4]
```

```python
plt.style.use('seaborn-white')
```

```python
fig, ax = plt.subplots(figsize=(8,6))

ax.set_title('Distribution of {} variances: Std. Normal Distribution'.format(num_vars), fontsize=15)

for sample_size in sample_sizes:
    variances = [np.random.normal(size=sample_size).var() for _ in range(num_vars)]

    sns.distplot(variances, kde=True, hist=False, ax=ax, label='Sample size: {}'.format(sample_size))

ax.vlines(x=1.0, ymin=0, ymax=32, linewidth=0.9, linestyle='--', label='Population variance')
    
ax.legend(fontsize=14)
fig.savefig('distribution-of-variances.png')
```
